```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x * x
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b*')
```

```python
fig = plt.figure()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
axes.plot(x, y)
axes2.plot(y, x, 'g--')
axes.set_xlabel('Eixo x')
axes.set_ylabel('Eixo y')
axes.set_title('Título')
```

```python
fig, ax = plt.subplots()
ax.plot(x, x ** 3, 'g--')
ax.plot(x, x**2, 'r-.')
```

```python
fig, ax = plt.subplots(2,2)
ax[0][0].plot(x, y, 'b--')
ax[0][1].plot(x, y*4, 'b--')
ax[1][0].plot(x, y**2, 'b--')
ax[1][1].plot(x, x**4, 'b--')
```

```python
fig, ax = plt.subplots(figsize=(6,10))
ax.plot(x, y, color='red')
ax.plot(x, x**4, 'g--')
```
