```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')

ax2 = fig.add_axes([0.1,0.4,0.4,0.4])
ax2.plot(x,z, color='green')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('title')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,z, color='purple')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')

ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
ax2.plot(x,z, color='green')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim([20,30])
ax2.set_ylim([0,3000])
ax2.set_title('zoom')
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,2))
ax[0].plot(x, y, 'r', ls='dashed', lw=2)
ax[1].plot(x, z, 'g-.')
```
